```python
# Simple plotting of Amazon Stock Price
# First Subplot
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(14,5))
ax1.plot(data["timestamp"], data["close"])
ax1.set_xlabel("Date", fontsize=12)
ax1.set_ylabel("Stock Price")
ax1.set_title("BCA Close Price History")

# Second Subplot
ax1.plot(data["timestamp"], data["high"], color="green")
ax1.set_xlabel("Date", fontsize=12)
ax1.set_ylabel("Stock Price")
ax1.set_title("BCA High Price History")

# Third Subplot
ax1.plot(data["timestamp"], data["low"], color="red")
ax1.set_xlabel("Date", fontsize=12)
ax1.set_ylabel("Stock Price")
ax1.set_title("BCA Low Price History")

# Fourth Subplot
ax2.plot(data["timestamp"], data["volume"], color="orange")
ax2.set_xlabel("Date", fontsize=12)
ax2.set_ylabel("Stock Price")
ax2.set_title("BCA's Volume History")
plt.show()
```

```python
m = Prophet()

# Drop the columns
prop_df = data.drop(['open', 'high', 'low','volume'], axis=1)
prop_df.rename(columns={'close': 'y', 'timestamp': 'ds'}, inplace=True)

prop_df.head()
```

```python
m = Prophet()
m.fit(prop_df)
```

```python
# Create Future dates
future_prices = m.make_future_dataframe(periods=365)

# Predict Prices
forecast = m.predict(future_prices)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
import matplotlib.dates as mdates

# Dates
starting_date = dt.datetime(2021, 4, 7)
starting_date1 = mdates.date2num(starting_date)
trend_date = dt.datetime(2021, 6, 7)
trend_date1 = mdates.date2num(trend_date)

pointing_arrow = dt.datetime(2021, 2, 18)
pointing_arrow1 = mdates.date2num(pointing_arrow)

# Learn more Prophet tomorrow and plot the forecast for amazon.
fig = m.plot(forecast)
ax1 = fig.add_subplot(111)
ax1.set_title("BCA Stock Price Forecast", fontsize=16)
ax1.set_xlabel("Date", fontsize=12)
ax1.set_ylabel("Close Price", fontsize=12)

# Forecast initialization arrow
ax1.annotate('Forecast \n Initialization', xy=(pointing_arrow1, 1350), xytext=(starting_date1,1700),
            arrowprops=dict(facecolor='#ff7f50', shrink=0.1),
            )

# Trend emphasis arrow
ax1.annotate('Upward Trend', xy=(trend_date1, 1225), xytext=(trend_date1,950),
            arrowprops=dict(facecolor='#6cff6c', shrink=0.1),
            )

ax1.axhline(y=1260, color='b', linestyle='-')

plt.show()
```

```python
fig2 = m.plot_components(forecast)
plt.show()
```

```python
# Monthly Data Predictions
m = Prophet(changepoint_prior_scale=0.01).fit(prop_df)
future = m.make_future_dataframe(periods=12, freq='M')
fcst = m.predict(future)
fig = m.plot(fcst)
plt.title("Monthly Prediction \n 1 year time frame")

plt.show()
```

```python
fig = m.plot_components(fcst)
plt.show()
```
